```python
import re
```

```python
def integer(text):
    a=re.findall("\d",text)
    b="".join(a)
    return int(b)
```

```python
df["price"]=df["price"].apply(integer)
```

```python
def panchuvation_remove(text):
    if text !=" ":
        a=re.findall("\w",text)
        b="".join(a)
        
    return b
```

```python
df=df.dropna()
```

```python
# df=df.iloc[:500,:]
```

```python
color=["Sunrise","Midnight","Purple","Starlight","Blue","Black","Red"]
```

```python
def remove_color(text):
    if text in color:
        return text
    else :
        return "White"
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vector=cv.fit_transform(df["tag"]).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
def recommend(company):
    index = df[df['company'] == company].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
#     print(distances)
    for i in distances[1:6]:
#         print(i)
        print(df.iloc[i[0],:])
```
